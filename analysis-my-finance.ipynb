```python
import csv
db = pd.read_csv('charge/2015analysis.csv',encoding='gbk')
len(db)
```

```python
data = db.drop(0)
data['Date'] = pd.to_datetime(data['Date'], format='%Y/%m/%d')
mdate=[]
for m in data.Date:
    months = pd.Period(m, freq='M')
    mdate.append(months)
mdate
data.Date = mdate
```

```python
data[:10]
```

```python
df_keys = ['Date', 'Category', 'Cost']
```

```python
df = pd.DataFrame(data, columns = df_keys)

df
```

```python
result = df.groupby(['Date']).Cost.sum()
result
```

```python
table_cost = df.pivot_table('Cost', index = 'Date', columns='Category', aggfunc=sum)
table_cost
```

```python
#result.plot(kind='barh')
table_cost.plot(subplots=True, layout=(3,3), kind='bar', figsize=(8,16))
plt.show()
```
