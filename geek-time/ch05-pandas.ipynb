```python
# practice
data = {'Chinese': [66, 95, 93, 90, 80, 80],
        'English': [65, 85, 92, 88, 90, 90],
        'Math': [None, 98, 96, 77, 90, 90]}
df = DataFrame(data, index=['张飞', '关羽', '赵云', '黄忠', '典韦', '典韦'])
df
```

```python
# 去重
df1 = df.drop_duplicates()
df1
```

```python
def cal_total(df):
    df['total'] = df['Chinese'] + df['Math'] + df['English']
    return df
df1 = df1.apply(cal_total, axis=1)
df1
```

```python
df1.isnull()
```

```python
# 用数学平均成绩填补张飞的值，不影响总体统计
df1['Math'].fillna(df['Math'].mean(), inplace=True)
```

```python
df1
```

```python
# 重新统计综合
df1 = df1.apply(cal_total, axis=1)
df1
```

```python
# total sort
df1.sort_values(['total'], ascending=[False], inplace=True)
df1
```
